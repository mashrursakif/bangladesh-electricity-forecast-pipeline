```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('electricity_data.csv')
df = df.iloc[1:]
df
```

```python
mask = df[df['Time'] == '24:00:00'].index
df.loc[mask, 'Time'] = '00:00:00'
```

```python
df
```

```python
df['date_time'] = df['Date'].astype(str) + ' ' + df['Time'].astype(str)
df['date_time'] = pd.to_datetime(df['date_time'], dayfirst=True)
df['Loadshed'] = df['Loadshed'].astype(int)
df['Generation(MW)'] = df['Generation(MW)'].astype(int)
df['Demand(MW)'] = df['Demand(MW)'].astype(int)
```

```python
plt.scatter(df['date_time'], df['Loadshed'], s=10)
```

```python
plt.scatter(df['date_time'], df['Generation(MW)'], s=10)
```

```python
plt.scatter(df['date_time'], df['Demand(MW)'], s=10)
```
