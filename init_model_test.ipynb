# Load Weather Data


```python
weather_df = pd.read_csv("BD_weather.csv")
weather_df
```

```python
weather_df["Date"] = pd.to_datetime(weather_df[["Year", "Month", "Day"]])
weather_df
```

```python
weather_df["Station"].unique()
```

```python
stations = [
    "Dhaka",
    "Chittagong",
    "Patuakhali",
    # 'Rangamati',
    # 'Bogra',
]
weather_df = weather_df[weather_df["Station"].isin(stations)]
```

```python
weather_df
```

```python
daily_weather_df = (
    weather_df.groupby("Date")
    .agg(
        {
            "Rainfall": "mean",
            "Sunshine": "mean",
            "Humidity": "mean",
            "Temperature": ["mean", "max", "min"],
        }
    )
    .reset_index()
)
daily_weather_df.columns = [
    "Date",
    "Rainfall",
    "Sunshine",
    "Humidity",
    "Temp_mean",
    "Temp_max",
    "Temp_min",
]
daily_weather_df
```

```python
power_df["Date"] = pd.to_datetime(power_df["Date"])
daily_power_df = (
    power_df[["Generation", "Demand", "Loadshed"]]
    .groupby(power_df["Date"])
    .sum()
    .reset_index()
)
```

```python
full_df = pd.merge(daily_power_df, daily_weather_df, on="Date")
full_df
```

```python
full_df["Year"] = full_df["Date"].dt.year
full_df["Month"] = full_df["Date"].dt.month
```

```python
full_df = full_df.sort_values(by="Date")
# full_df['loadshed_prev'] = full_df['Loadshed'].shift(1)
full_df["generation_prev"] = full_df["Generation"].shift(1)
```

```python
full_df
```

```python
full_df["year_month"] = full_df["Year"].astype(str) + "_" + full_df["Month"].astype(str)
full_df["year_month"] = full_df["year_month"].astype("category")
```

# LGBM


```python
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
target_col = "Generation"
X = full_df.drop(columns=["Generation", "Demand", "Loadshed", "Date"])
y = full_df[target_col]

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42)

train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=["year_month"])
valid_data = lgb.Dataset(
    X_valid, label=y_valid, reference=train_data, categorical_feature=["year_month"]
)
```

```python
params = {
    "objective": "regression",
    "metric": "rmse",
    "boosting_type": "gbdt",
    "num_boost_round": 1000,
    "learning_rate": 1e-2,
    "num_leaves": 63,
    # 'max_depth': 2,
    "verbose": -1,
    "seed": 42,
}

model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, valid_data],
    callbacks=[lgb.early_stopping(stopping_rounds=500), lgb.log_evaluation(period=100)],
)
```

```python
valid_preds = model.predict(X_valid)
print("RMSE ", np.sqrt(mean_squared_error(y_valid, valid_preds)))
print("MAE ", mean_absolute_error(y_valid, valid_preds))
```

```python
importance = model.feature_importance()
features = model.feature_name()

df_importance = pd.DataFrame(
    {"Feature": features, "Importance": importance}
).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(df_importance["Feature"], df_importance["Importance"])
plt.gca().invert_yaxis()
plt.xlabel("Importance")
plt.title("Feature Importance")
plt.tight_layout()
plt.show()
```

```python
valid_df = full_df.loc[X_valid.index].copy()
valid_df = valid_df.sort_values(by="Date")
valid_df["Pred"] = model.predict(valid_df[X_train.columns])
plt.figure(figsize=(20, 5))
plt.plot(
    valid_df["Date"],
    valid_df[target_col],
    color="red",
    marker="o",
    markersize=2,
    label="Target",
)
plt.plot(
    valid_df["Date"],
    valid_df["Pred"],
    color="blue",
    marker="x",
    markersize=2,
    label="Prediction",
)
plt.legend()
```

```python
view_df = full_df.copy()
view_df["Pred"] = model.predict(full_df[X_train.columns])
plt.figure(figsize=(20, 5))
plt.plot(view_df["Date"], view_df[target_col], color="red", marker="o", markersize=2)
plt.plot(view_df["Date"], view_df["Pred"], color="blue", marker="x", markersize=2)
```

```python
train_df = full_df[full_df["Year"] == 2022]
valid_df = full_df[full_df["Year"] == 2023]
X_train = train_df.drop(columns=["Generation", "Demand", "Loadshed", "Date"])
y_train = train_df[target_col]
X_valid = valid_df.drop(columns=["Generation", "Demand", "Loadshed", "Date"])
y_valid = valid_df[target_col]

train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=["year_month"])
valid_data = lgb.Dataset(
    X_valid, label=y_valid, reference=train_data, categorical_feature=["year_month"]
)
```

```python
params = {
    "objective": "regression",
    "metric": "rmse",
    "boosting_type": "gbdt",
    "num_boost_round": 1000,
    "learning_rate": 1e-2,
    "num_leaves": 63,
    # 'max_depth': 2,
    "verbose": -1,
}

model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, valid_data],
    callbacks=[lgb.early_stopping(stopping_rounds=500), lgb.log_evaluation(period=200)],
)
```
